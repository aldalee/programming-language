```python
# 对于几个公司的财经数据，如何构造公司代码和股票价格的字典？
If = [('AXP', 'American Express Company', '78.51'),
      ('BA', 'The Boeing Company', '184.76'),
      ('CAT', 'Caterpillar Inc.', '96.39'),
      ('CSCO', 'Cisco System,Inc.', '33.71'),
      ('CVX', 'Chevron Corporation', '106.09')]
```

```python
# 方法一
CompanyCode = []
SharePrice = []
for i in range(len(If)):
    CompanyCode.append(If[i][0])
    SharePrice.append(If[i][2])
CS = dict(zip(CompanyCode, SharePrice))
```

```python
CS
```

```python
# 方法二
CS = {}
for item in If:
    CS[item[0]] = item[2]
```

```python
CS
```

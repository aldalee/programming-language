# 数据可视化

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## 折线图

```python
plt.plot([3, 4, 7, 6, 2, 8, 9])
```

```python
plt.plot([3, 4, 7, 6, 2, 8, 9], 'g--')
```

```python
plt.plot(range(7), [3, 4, 7, 6, 2, 8, 9], 'rD')
```

```python
t = np.arange(0, 4, 0.1)
```

```python
plt.plot(t, t, t, t + 2, t, t ** 2)
```

## 散点图

```python
plt.scatter(range(7), [3, 4, 7, 6, 2, 8, 9])
```

```python
plt.figure(figsize=(8, 6), dpi=100)
t = np.arange(-20, 20, 0.1)
plt.plot(t, t * np.log(t), color='red', linestyle='-', linewidth=2, label='Line 1')
plt.plot(t, np.sin(t) * t ** 2, color='green', linestyle=':', label='Line 2')
plt.plot(t, np.cos(t) * t ** 2, color='blue', linestyle='-.', label='Line 3')
plt.plot(t, np.tan(t), color='purple', linestyle='-.', label='Line 4')
plt.legend(loc='best')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Plot Example')
```

## 柱状图

```python
plt.bar(range(7), [3, 4, 7, 6, 2, 8, 9])
```

## 饼图

```python
plt.pie([3, 4, 7, 6, 2, 8, 9], labels=['a', 'b', 'c', 'd', 'd', 'f', 'g'])
```

## 子图 subplot和subplots

```python
x = np.linspace(-np.pi, np.pi, 300)
plt.figure(1)
plt.subplot(211)
plt.plot(x, np.sin(x), color='r')
plt.subplot(212)
plt.plot(x, np.cos(x), color='g')
```

```python
x = np.linspace(-np.pi, np.pi, 300)
fig, (ax0, ax1) = plt.subplots(2, 1)
ax0.plot(x, np.sin(x), color='r')
ax0.set_title('sin(x)')
plt.subplots_adjust(hspace=0.5)
ax1.plot(x, np.cos(x), color='g')
ax1.set_title('cos(x)')
```

## 子图 axes()
axes([left,bottom,width,height]) 参数范围为(0,1)

```python
x = np.linspace(-np.pi, np.pi, 300)
plt.axes([0.1, 0.1, 0.9, 0.9])
plt.plot(x, np.sin(x), color='r')
plt.axes([0.58, 0.2, 0.4, 0.3])
plt.plot(x, np.cos(x), color='b')
```

## pandas绘图

```python
data = {'A': np.random.randn(100) + 2,
        'B': np.random.randn(100),
        'C': np.random.randn(100) - 2,
        'D': np.random.randn(100) + 3}
df = pd.DataFrame(data, columns = ['A', 'B', 'C', 'D'])
#使用diff绘制
df.diff().hist(color="r", alpha=0.5, bins=15)
```

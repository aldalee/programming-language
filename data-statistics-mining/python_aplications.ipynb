# 古滕堡项目

```python
from nltk.corpus import gutenberg
```

```python
gutenberg.fileids()
```

```python
all_words = gutenberg.words('shakespeare-hamlet.txt')
```

```python
len(all_words)
```

```python
len(set(all_words))
```

```python
all_words.count('Hamlet')
```

```python
long_words = [w for w in set(all_words) if len(w) > 12]
```

```python
sorted(long_words)
```

```python
from nltk.probability import *
```

```python
fd2 = FreqDist([sx.lower() for sx in all_words if sx.isalpha()])
```

```python
fd2.B()
```

```python
fd2.N()
```

```python
fd2.tabulate(20)
```

```python
fd2.plot(20)
```

```python
fd2.plot(20, cumulative=True)
```

# 就职演说语料库

```python
from nltk.corpus import inaugural
```

```python
fd3 = FreqDist([s for s in inaugural.words()])
```

```python
fd3.freq('freedom')
```

```python
cfd = ConditionalFreqDist(
    (fileid, len(w))
    for fileid in inaugural.fileids()
    for w in inaugural.words(fileid)
    if '1980' < fileid < '2010')
```

```python
cfd.items()
```

```python
cfd.plot()
```
